```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from pprint import pprint

# Import API key
from config import weather_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# create empty lists to hold data for each successful API call
final_cities = []
final_countries = []
final_lats = []
final_lngs = []
temps = []
cloudiness = []
humidity = []
wind_speed = []

test_cities = []
x = 0
while x < 122:
    test_cities.append(cities[x])
    x += 1

base_url = "http://api.openweathermap.org/data/2.5/weather?"

print("Beginning Data Retrieval\n-----------------------")
x = 1
for city in cities:
    # Build query url
    query_url = base_url + "appid=" + weather_key + "&units=imperial&q=" + city
    
    # Get Weather Data
    weather_response = requests.get(query_url)
    weather_json = weather_response.json()
    
    # try to append city dictionary to list
    try:
        print(f'Processing Record {x} | {city}')
        final_cities.append(weather_json['name'])
        final_countries.append(weather_json['sys']['country'])
        final_lats.append(weather_json['coord']['lat'])
        final_lngs.append(weather_json['coord']['lon'])
        temps.append(weather_json['main']['temp'])
        humidity.append(weather_json['main']['humidity'])
        wind_speed.append(weather_json['wind']['speed'])
        cloudiness.append(weather_json['clouds']['all'])
        x += 1
    except KeyError:
        print(f'{city} not found... skipping')
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# create a Data Frame from city name, lat, long, and temp
weather_dictionary = {
    "City" : final_cities,
    "Country" : final_countries,
    "Lat" : final_lats,
    "Lng" : final_lngs,
    "Temp" : temps,
    "Cloudiness" : cloudiness,
    "Humidity" : humidity,
    "Wind Speed" : wind_speed
}
weather_data = pd.DataFrame(weather_dictionary)
weather_data.to_csv("weather_output.csv", index=False)
weather_data.head()
```

## Linear Regression

```python
# create northern hemisphere DataFrame
northern_df = weather_data.loc[weather_data["Lat"] >= 0, :]

# create southern hemisphere DataFrame
southern_df = weather_data.loc[weather_data["Lat"] < 0, :]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# perform linear regression on norther hemisphere lat vs temp and plot
nh_slope, nh_intercept, nh_rval, nh_pval, nh_std_err = linregress(northern_df["Lat"], northern_df["Temp"])
plt.scatter(northern_df["Lat"], northern_df["Temp"])
plt.xlabel("Latitude")
plt.ylabel("Temperature")
plt.title("Norther Hemisphere Temperature vs Latitude")
line_eq = "y = " + str(round(nh_slope,2)) + "x + " + str(round(nh_intercept,2))
regress_values = northern_df["Lat"] * nh_slope + nh_intercept
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.plot(northern_df["Lat"], regress_values,"r-")
print(f"The r-value is {nh_rval}")
# The linear regression here is demonstrating a strong negative correlation between latitude and temperature in the
# northern hemisphere--- as latitude numbers increase (ie, you travel away from the equator), temperature decreases
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# perform linear regression on southern hemisphere lat vs temp and plot
sh_slope, sh_intercept, sh_rval, sh_pval, sh_std_err = linregress(southern_df["Lat"], southern_df["Temp"])
plt.scatter(southern_df["Lat"], southern_df["Temp"])
plt.xlabel("Latitude")
plt.ylabel("Temperature")
plt.title("Southern Hemisphere Temperature vs Latitude")
s_line_eq = "y = " + str(round(sh_slope,2)) + "x + " + str(round(sh_intercept,2))
regress_values = southern_df["Lat"] * sh_slope + sh_intercept
plt.plot(southern_df["Lat"], regress_values,"r-")
plt.annotate(s_line_eq,(-50,40),fontsize=15,color="red")
plt.show()
print(f"The r-value is {sh_rval}")
# The linear regression here is demonstrating a strong negative correlation between latitude and temperature in the
# southern hemisphere--- as latitude numbers increase (ie, you travel closer to the equator), temperature increases
```
